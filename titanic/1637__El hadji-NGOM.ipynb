```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import IPython
%matplotlib inline
```

```python
# a)
plt.interactive(True)
df = pd.read_csv('data/UN.csv')
df.keys()
```

```python
# c)
a = df['GDPperCapita'].values
b = df['infantMortality'].values
print len(a), len(b)
```

```python
plt.figure(figsize=(12,7))
plt.scatter(a,b,c='r')
plt.xlim(0.0, 50000)
plt.ylim(0.0, 200)
plt.grid()
plt.title('GDPperCapita vs infantMortality')
plt.ylabel('infantMortality')
plt.xlabel('GDPperCapita') 
plt.legend(['mortality/GDP'])
plt.draw()
plt.draw()
```
