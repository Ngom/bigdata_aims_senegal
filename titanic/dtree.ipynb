First, we will load the data into an array of dictionaries. This load_data function can be used to load both training and test data.

See:  http://scikit-learn.org/stable/modules/feature_extraction.html#dict-feature-extraction

```python
import numpy as np
import csv
def load_data(path, is_training):
    """
    is_training = 
        True:  Return raw Training Set and Labels
        False: Return raw Test Set
    """
    with open(path) as csvfile:
        training_file = csv.DictReader(csvfile)
        data = []
        labels = []
        for row in training_file:
            if is_training:
                labels.append(row['Survived'])
                del row['Survived'] # we can't train on survived!
            del row['Name'] # for simplicity, we won't try to analyze name, ticket, or cabin
            del row['Ticket']
            del row['Cabin']
            try:
                row['Fare'] = min(250., float(row['Fare'])) # And lets convert strings to floats/ints as makes sense
            except ValueError:
                row['Fare'] = 20.
            try:
                row['Age'] = float(row['Age'])
            except ValueError:
                row['Age'] = 30. # assign an "average" age for now, (assigning "None" causes errors)
            row['sibsp'] = int(row['SibSp'])
            row['parch'] = int(row['Parch'])
            row['pclass'] = int(row['Pclass'])
            row['Embarked'] = row['Embarked'] if row['Embarked'] else 'S'
            data.append(row)
        if is_training:    
            return np.array(data), np.array(labels)
        else:
            return np.array(data)

training_data, labels = load_data('data/train.csv', True)    
print training_data[0], labels[0]
```

Now we will try taking a look at the feature distribution, by using Python's matplotlib.

```python
# Used so that when we call plt.show(), the graphs show up inside iPython Notebook
%pylab inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# to get subplots side by side: http://stackoverflow.com/a/11404223/264003

def plot_distributions(data):
    continuous_columns = filter(lambda x: isinstance(data[0][x], float), data[0].keys())
    discrete_columns = filter(lambda x: not isinstance(data[0][x], float), data[0].keys())
    
    print "Continuous Features: {}".format(continuous_columns)
    n_rows = (len(continuous_columns) + 2 //2) // 2 # ceiling division trick
    plt.figure(figsize=(20, 8*n_rows), dpi=100)
    for i, column in enumerate(continuous_columns):
        rows = np.array([row[column] for row in data])   
        plt.subplot(n_rows, 2, i+1)
        plt.hist(rows, 50)
        plt.title(column)
    plt.show()
   
    print "Discrete Features: {}".format(discrete_columns) 
    n_rows = (len(discrete_columns) + 2 //2) // 2
    plt.figure(figsize=(20, 8*n_rows), dpi=100)
    for i, column in enumerate(discrete_columns):
        counts = Counter(row[column] for row in data)
        index = np.arange(len(counts))
        plt.subplot(n_rows, 2, i+1)
        plt.bar(index, counts.values())
        plt.xticks(index, counts.keys())
        plt.title(column)
    plt.show()

plot_distributions(training_data)
```

```python
#now lets try some feature extraction
from sklearn.feature_extraction import DictVectorizer
vectorizer = DictVectorizer()
vectorizer.fit(training_data)
cleaned_training_data = vectorizer.transform(training_data).toarray()

print training_data[0], labels[0]
print vectorizer.get_feature_names()
print cleaned_training_data[0], labels[0]
# The output shows how `embarked` and `sex` have been transformed into One-Hot encoding.
```

```python
# Now lets build a decision tree
from sklearn import tree
classifier = tree.DecisionTreeClassifier(min_samples_leaf=5, max_depth=4)
classifier.fit(cleaned_training_data, labels)
# let's see how our decision tree works agains the training data (should be near-perfect)
print "First 5 training labels"
print labels[0:5]
print "Prediction on first 5 training Rows"
print classifier.predict(cleaned_training_data[0:5])
```

```python
# Now lets try running it on the test set, and take a look at the first 5 rows

test_data = load_data('data/test.csv', False) # Load test set
print test_data[0]

test_data = vectorizer.transform(test_data).toarray() # Vectorize it
print test_data[0]

predictions = classifier.predict(test_data)
print "Prediction on first 5 test rows"
print predictions[0:5]
```

Let's learn about how to transform categorical data into something that scikit-learn can read (One-Hot encoding). 

This estimator transforms each categorical feature with m possible values into m binary features, with only one active. THUS, each categorical feature blows up into m more features, increasing the size of your feature vector by a lot.

see: http://scikit-learn.org/stable/modules/preprocessing.html#encoding-categorical-features

```python
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()
enc.fit([[0],[1],[2]]) # Assume we have a data set of Red, Yello, Green, like a stoplight

#Lets transform a Red (the first) into One-Hot encoding
enc.transform([[0]]).toarray()
```
